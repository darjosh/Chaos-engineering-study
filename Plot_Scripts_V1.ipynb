```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, RegularPolygon
from matplotlib.path import Path
from matplotlib.projections.polar import PolarAxes
from matplotlib.projections import register_projection
from matplotlib.spines import Spine
from matplotlib.transforms import Affine2D

def radar_factory(num_vars, frame='circle'):
    theta = np.linspace(0, 2 * np.pi, num_vars, endpoint=False)

    class RadarTransform(PolarAxes.PolarTransform):
        def transform_path_non_affine(self, path):
            if path._interpolation_steps > 1:
                path = path.interpolated(num_vars)
            return Path(self.transform(path.vertices), path.codes)

    class RadarAxes(PolarAxes):
        name = 'radar'
        PolarTransform = RadarTransform

        def __init__(self, *args, **kwargs):
            super().__init__(*args, **kwargs)
            self.set_theta_zero_location('N')

        def fill(self, *args, closed=True, **kwargs):
            return super().fill(closed=closed, *args, **kwargs)

        def plot(self, *args, **kwargs):
            lines = super().plot(*args, **kwargs)
            for line in lines:
                self._close_line(line)

        def _close_line(self, line):
            x, y = line.get_data()
            if x[0] != x[-1]:
                x = np.concatenate((x, [x[0]]))
                y = np.concatenate((y, [y[0]]))
                line.set_data(x, y)

        def set_varlabels(self, labels, paddings=None, fontsize=36):
            self.set_thetagrids(np.degrees(theta), labels)
            for i, label in enumerate(self.get_xticklabels()):
                label.set_fontsize(fontsize)
                label.set_verticalalignment('center')
                if paddings:
                    # Apply individual padding
                    radial_position = label.get_position()[1] - (paddings[i] / 100.0)
                    label.set_y(radial_position)

        def _gen_axes_patch(self):
            if frame == 'circle':
                return Circle((0.5, 0.5), 0.5)
            elif frame == 'polygon':
                return RegularPolygon((0.5, 0.5), num_vars, radius=0.5, edgecolor="k")
            else:
                raise ValueError("Unknown value for 'frame': %s" % frame)

        def draw(self, renderer):
            if frame == 'polygon':
                gridlines = self.yaxis.get_gridlines()
                for gl in gridlines:
                    gl.get_path()._interpolation_steps = num_vars
            super().draw(renderer)

        def _gen_axes_spines(self):
            if frame == 'circle':
                return super()._gen_axes_spines()
            elif frame == 'polygon':
                spine = Spine(axes=self,
                              spine_type='circle',
                              path=Path.unit_regular_polygon(num_vars))
                spine.set_transform(Affine2D().scale(0.5).translate(0.5, 0.5)
                                    + self.transAxes)
                return {'polar': spine}
            else:
                raise ValueError("Unknown value for 'frame': %s" % frame)

    register_projection(RadarAxes)
    return theta

# Your Data
spoke_labels = ['0-10', '11-50', '51-100', '101-500', '501-1000', '>1000']
data_values = [31, 34, 16, 34, 16, 25]

# Number of variables
N = len(spoke_labels)
theta = radar_factory(N, frame='polygon')

# Define grid levels
max_value = max(data_values)
grid_values = np.linspace(0, max_value + 20, 6)  # Ensure 6 grid lines with room for the largest value

# Create radar chart
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(projection='radar'))
fig.subplots_adjust(top=0.85, bottom=0.05)
ax.set_rgrids(grid_values)  # Set grid levels

# Ensure grid lines are visible
ax.yaxis.grid(True, linestyle='--', color='gray', alpha=0.7)

# Remove radial grid labels
ax.set_yticklabels([])

# Add title
ax.set_title("Chaos Monkey", loc='center', fontsize=36, pad=24)


# Define individual padding for data point annotations
data_point_padding = [-4, -5, 5, -4, 5, 5]   # Padding for each point's annotation

# Define individual padding for spoke labels
spoke_label_padding = [3, 20, 25, 5, 35, 20]  # Padding for each spoke label

# Plot data
ax.plot(theta, data_values)  # Plot data points with a line

# Annotate data points with a rectangular box and individual padding
for i, val in enumerate(data_values):
    ax.text(theta[i], val + data_point_padding[i], f'{val}', ha='center', va='center', fontsize=36, color='black',
            bbox=dict(boxstyle="round,pad=0.2", edgecolor='grey', facecolor='white'))

# Set spoke labels with individual padding
ax.set_varlabels(spoke_labels, paddings=spoke_label_padding, fontsize=36)

plt.tight_layout()
#plt.savefig("Chaos.png", dpi=300)  # Save as PNG with 300 DPI
plt.savefig("CE_Commits_ChaosMonkey.pdf", dpi=300)  # Save as PDF with 300 DPI

# Optionally, display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Updated data for the chart
sizes = [90, 27, 43, 648, 415]
labels = ['Teaching', 'Learning', 'Research', 'Development', 'Unclassified']

# Calculate percentages dynamically
percentages = [
    round(size / sum(sizes) * 100, 1) for size in sizes
]

# Create labels with category and percentage only
# Exempt 'Teaching', 'Learning', and 'Research' from breaking into new lines
label_with_percent = [
    f"{labels[i]} {percentages[i]}%" if labels[i] in ['Teaching', 'Learning', 'Research']
    else f"{labels[i]}\n{percentages[i]}%" for i in range(len(labels))
]

# Define readable colors for the pie chart
colors = [
    "#ff9999",  # Light red
    "#66b3ff",  # Light blue
    "#99ff99",  # Light green
    "#ffcc99",  # Light orange
    "#c2c2f0"   # Light purple
]

# Create the pie chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts = ax.pie(
    sizes,
    labels=label_with_percent,
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor': 'white'},
    labeldistance=1.1
)

# Define individual padding and rotation for each label
padding = [0.5, 0.5, 1.1, 1.0, 1.20]
custom_rotations = [1, 5, -1, -32, -50]

# Adjust the position and rotation of each label
for i, text in enumerate(texts):
    x, y = text.get_position()
    angle = np.arctan2(y, x)
    distance = padding[i]
    new_x = np.cos(angle) * distance
    new_y = np.sin(angle) * distance
    text.set_position((new_x, new_y))
    text.set_rotation(custom_rotations[i])
    text.set_fontsize(28)  # Adjusted font size for multi-line labels
    text.set_ha('center')

    # Set all text colors to black
    text.set_color("black")

# Additional manual adjustments for label positioning
texts[0].set_position((texts[0].get_position()[0] + 0.28, texts[0].get_position()[1] + 0.61))

texts[1].set_position((
    texts[1].get_position()[0] - 1.09,  # Move left
    texts[1].get_position()[1] + 0.26   # Move up
))

texts[2].set_position((
    texts[2].get_position()[0] - 0.99,  # Move left
    texts[2].get_position()[1] + 0.28   # Move up
))

texts[2].set_position((
    texts[2].get_position()[0] + 0.57,  # Move right
    texts[2].get_position()[1] - 0.19   # Move down
))

texts[3].set_position((
    texts[3].get_position()[0] - 0.55,  # Move to the left
    texts[3].get_position()[1] + 0.25   # Move upward
))

texts[3].set_position((
    texts[3].get_position()[0] + 0.85,  # Move to the right
    texts[3].get_position()[1] + 0.32   # Move downward
))

texts[4].set_position((
    texts[4].get_position()[0] - 0.55,  # Move to the left
    texts[4].get_position()[1] - 0.42   # Move upward
))

# Set chart title
#plt.title("Commits in Categorized Projects", fontsize=35, pad=10)

plt.tight_layout()
# plt.savefig("Chaos.png", dpi=300)  # Save as PNG with 300 DPI
plt.savefig("CE_Projects_By_Type.pdf", dpi=300)  # Save as PDF with 300 DPI

# Show the chart
plt.show()
```
